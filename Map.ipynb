```python
import folium
```

## Map of NYC

```python
lat = 40.723663772
lon = -73.989829374
```

```python
mapa = folium.Map(location = [lat, lon], zoom_start=13)
mapa
```

## Requirements

```python
#0. Oficina:

of_lat = 40.726155
of_lon = -73.995625

#1. Aeropuerto:

air_lat = 40.732918
air_lon = -73.995468

#2. Nurseries:

nur_lat = 40.72584403329508
nur_lon = -73.99665355682373

#3. Hairdresser:

hd_lat = 40.740808060785625
hd_lon = -74.00659656331318

#4. Starbucks:

st_lat = 40.72994333023035
st_lon = -73.9939131949986
 
#5. Nightclub:

nc_lat = 40.72186578600761
nc_lon = -73.98048518308426
```

## Icons in the map

```python
icono0 = Icon(color = "orange",
             prefix = "fa",
             icon = "briefcase",
             icon_color = "black",
)
loc = {"location":[of_lat,of_lon],
      "tooltip": "The Office"}

marker_of = Marker(**loc, icon = icono0)
marker_of.add_to(mapa)

icono1 = Icon(color = "blue",
             prefix = "fa",
             icon = "plane",
             icon_color = "black",
)
loc = {"location":[air_lat,air_lon],
      "tooltip": "Airport"}

marker_air = Marker(**loc, icon = icono1)
marker_air.add_to(mapa)

icono2 = Icon(color = "cadetblue",
             prefix = "fa",
             icon = "baby",
             icon_color = "black",
)
loc = {"location":[nur_lat,nur_lon],
      "tooltip": "Nursery"}

marker_nur = Marker(**loc, icon = icono2)
marker_nur.add_to(mapa)

icono3 = Icon(color = "pink",
             prefix = "fa",
             icon = "cut",
             icon_color = "black",
)
loc = {"location":[hd_lat,hd_lon],
      "tooltip": "Hairdresser"}

marker_hd = Marker(**loc, icon = icono3)
marker_hd.add_to(mapa)


icono4 = Icon(color = "green",
             prefix = "fa",
             icon = "coffee",
             icon_color = "black",
)
loc = {"location":[st_lat,st_lon],
      "tooltip": "Starbucks"}

marker_st = Marker(**loc, icon = icono4)
marker_st.add_to(mapa)

icono5 = Icon(color = "red",
             prefix = "fa",
             icon = "glass",
             icon_color = "black",
)
loc = {"location":[nc_lat,nc_lon],
      "tooltip": "Night Club"}

marker_nc = Marker(**loc, icon = icono5)
marker_nc.add_to(mapa)
```
